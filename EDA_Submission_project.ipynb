### Handling missing values

````python
# check data type of last review
print(data["last_review"].dtypes)
````

````python
# if data is numerical check mean or median value
print("mean of reviews_per_month",data["reviews_per_month"].mean())
print("median of reviews_per_month",data["reviews_per_month"].median())

# for categorical data go for mode
print("mode of name",data["name"].mode())

print("mode of host_name",data["host_name"].mode())
````

````python
#  now plot graphs to check if data is skewed or not
plt.figure(figsize = (5,3))
sns.histplot(data = data["reviews_per_month"],kde=True)
````

````python
# check skewness of review per month
print("skewness of review_per_month",data["reviews_per_month"].skew())
````

````python
# now replacing missing values with mean,median and mode
# right skewed data = median
# left skewed data = median
# normal distributed data = mean
# categorical data = mode
````

````python
# replacing missing values
data.fillna({
    "reviews_per_month": data["reviews_per_month"].median(),
    "name": data["name"].mode()[0],
    "host_name": data["host_name"].mode()[0],
    "last_review": data["last_review"].min()
}, inplace=True)
````
